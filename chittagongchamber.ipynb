```python
from PyPDF2 import PdfReader
```

```python

reader = PdfReader('chittagongchamber.pdf')
print(len(reader.pages))

# Extract text from the first page
page = reader.pages[0]
print(page.extract_text())
```

```python
for i in range(len(reader.pages)):
    page =reader.pages[i]
    print(page.extract_text())
    
```
