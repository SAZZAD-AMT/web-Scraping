```python
from PyPDF2 import PdfReader
import re
```

```python
# Load the PDF
reader = PdfReader('chittagongchamber.pdf')
```

```python


def extract_data(text):
    """
    Extracts company name, address, and phone number from the given text.
    
    Args:
        text (str): The input text to process.
    
    Returns:
        list: A list of dictionaries containing company details.
    """
    data = []  # To store extracted data
    lines = text.split('\n')  # Split text by lines
    company_pattern = re.compile(r'^\d+\s+([A-Za-z0-9 &.,]+)')  # Pattern for company names

    for i, line in enumerate(lines):
        line = line.strip()  # Clean up the line
        match = company_pattern.match(line)
        if match:  # If the line matches the company name pattern
            company = match.group(1).strip()
            address = ''
            phone = 'N/A'
            
            # Look for address and phone in subsequent lines
            for j in range(i + 1, len(lines)):
                next_line = lines[j].strip()
                if company_pattern.match(next_line):  # Stop at the next company entry
                    break
                if re.match(r'^01[3-9]\d{8}$', next_line):  # Detect Bangladeshi phone numbers
                    phone = next_line
                else:
                    address += next_line + ' '  # Append to address
            
            # Append to the data list
            data.append({
                'Company': company,
                'Address': address.strip(),
                'Phone': phone
            })

    return data
```

```python
# Extract text from all pages
all_text = ''
for page in reader.pages:
    all_text += page.extract_text()

# Extract data
data = extract_data(all_text)

# Print the results
for entry in data:
    print(f"Company: {entry['Company']}")
    print(f"Address: {entry['Address']}")
    print(f"Phone: {entry['Phone']}")
    print("-" * 40)
```
